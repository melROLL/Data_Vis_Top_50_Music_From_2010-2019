```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas dataframe
df = pd.read_csv('top50MusicFrom2010-2019.csv')

# Group the dataframe by genre and count the number of songs in each genre
genre_counts = df.groupby('the genre of the track')['title'].count()

# Sort the resulting series in descending order
genre_counts = genre_counts.sort_values(ascending=False)

# Select the top 5 genres
top_10_genres = genre_counts.head(5)

# Create a bar plot of the top 5 genres
top_10_genres.plot(kind='bar', color=['black'])

# Set the plot title and axis labels
plt.title('Top 5 Music Genres by Number of Songs in the top 50 between 2010 and 2019')
plt.xlabel('Genre')
plt.ylabel('Number of Songs')

# Show the plot
plt.show()

# Select the top 30 genres
top_10_genres = genre_counts.head(30)

# Create a bar plot of the top 30 genres
top_10_genres.plot(kind='bar', color=['black'])

# Set the plot title and axis labels
plt.title('Top 30 Music Genres by Number of Songs in the top 50 between 2010 and 2019')
plt.xlabel('Genre')
plt.ylabel('Number of Songs')

# Show the plot
plt.show()

# Select the top 30 genres
top_10_genres = genre_counts.head(60)

# Create a bar plot of the top 30 genres
top_10_genres.plot(kind='bar', color=['black'])

# Set the plot title and axis labels
plt.title('Top 60 Music Genres by Number of Songs in the top 50 between 2010 and 2019')
plt.xlabel('Genre')
plt.ylabel('Number of Songs')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the dataset
df = pd.read_csv('top50MusicFrom2010-2019.csv')

# Get statistical measures of 'Beats.Per.Minute'
beats_per_min_stats = df['Beats.Per.Minute -The tempo of the song'].describe()

# Print the statistics
print(beats_per_min_stats)

# Plot a histogram of 'Beats.Per.Minute'
sns.histplot(data=df, x='Beats.Per.Minute -The tempo of the song', bins=20)

# Show the plot
plt.show()


# Plot a histogram of 'Beats.Per.Minute'
sns.histplot(data=df, x='Beats.Per.Minute -The tempo of the song', bins=20, color='black')

# Set the x and y labels
plt.xlabel('Beats Per Minute')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas dataframe
df = pd.read_csv('top50MusicFrom2010-2019.csv')

# Group the dataframe by artist and count the number of songs for each artist
artist_counts = df.groupby('artist')['title'].count()

# Sort the resulting series in descending order
artist_counts = artist_counts.sort_values(ascending=False)

# Select the top 5 artists
top_5_artists = artist_counts.head(5)

# Create a bar plot of the top 5 artists
top_5_artists.plot(kind='bar', color=['black'], width=0.2)

# Set the plot title and axis labels
plt.title('Top 5 Artists by Number of Songs')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')

# Show the plot
plt.show()

# Select the top 10 artists
top_5_artists = artist_counts.head(10)

# Create a bar plot of the top 10 artists
top_5_artists.plot(kind='bar', color=['black'], width=0.2)

# Set the plot title and axis labels
plt.title('Top 10 Artists by Number of Songs')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')

# Show the plot
plt.show()


print(top_5_artists)
```
